#**Which age has the highest total purchase amount based on the sum of all purchases?**

#**What is the distribution of shipping types (Express vs Free) across different age groups?**

```python
# We make all the values lowercase and clean them
df['Shipping Type'] = df['Shipping Type'].str.strip().str.lower()

# We standardize the names
df['Shipping Type'] = df['Shipping Type'].replace({
    'express shipping': 'express',
    'express delivery': 'express',
    'free shipping': 'free',
    'free delivery': 'free'
})

# Dividing ages into categories
bins = [18, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculating the percentage of each shipping type within each age group
age_shipping_percent = (
    df.groupby(['Age Group', 'Shipping Type']).size()
      .groupby(level=0)
      .apply(lambda x: 100 * x / x.sum())
      .unstack(fill_value=0)
)
# We arrange the columns to ensure that they are 'express' and 'free'
age_shipping_percent = age_shipping_percent[['express', 'free']]

plt.figure(figsize=(12,7))
bar_plot = age_shipping_percent.plot(
    kind='bar',
    figsize=(12,7),
    color=['#4CAF50', '#FF9800'],  # Express = برتقالي / Free = أخضر
    edgecolor='black',
    width=0.65
)

plt.title('Shipping Type Distribution by Age Group (%)', fontsize=20, fontweight='bold', pad=15)
plt.xlabel('Age Group', fontsize=14, fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=14, fontweight='bold')
plt.xticks(rotation=0, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.legend(title='Shipping Type', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the percentages above the columns in a well-organized manner and at an appropriate distance.
for container in plt.gca().containers:
    plt.bar_label(
        container,
        fmt='%.1f%%',
        label_type='edge',
        fontsize=11,
        fontweight='bold',
        padding=3  # Distance between the number and the column
    )

plt.tight_layout()
plt.show()
```

#**How does the total purchase amount and the number of customers differ between males and females?**

```python
gender_sales = df.groupby('Gender')['Purchase Amount (USD)'].sum()
gender_sales
```

```python
gender_counts = df['Gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(20, 6))

# The first chart (Total Purchases)
plt.subplot(1, 2, 1)
plt.pie(
    gender_sales,
    labels=gender_sales.index,
    explode=[0.04, 0],
    colors=['pink', 'blue'],
    autopct="%0.2f%%",
    startangle=30,
    shadow=True
)
plt.title('Total Purchase Amount by Gender', fontsize=16, fontweight='bold')

# The second chart (Number of buyers)
plt.subplot(1, 2, 2)
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%0.1f%%',
    startangle=140,
    colors=['blue', 'pink'],
    explode=(0.03, 0.03),
    shadow=True
)
plt.title('Number of Customers by Gender', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

#**Which product category had the highest total sales amount?**

```python
# We calculate the total sales for each category
category_sales = df.groupby('Category', as_index=False)['Purchase Amount (USD)'].sum()

# We arrange them in descending order
category_sales = category_sales.sort_values(by='Purchase Amount (USD)', ascending=False)

plt.figure(figsize=(12,7))
sns.barplot(
    data=category_sales,
    x='Category',
    y='Purchase Amount (USD)',
    hue='Category',
    palette='viridis'
)

plt.title('Total Sales by Category', fontsize=18, fontweight='bold', pad=15)
plt.xlabel('Category', fontsize=14, fontweight='bold')
plt.ylabel('Total Purchase Amount (USD)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

# Add values above the columns
for i, v in enumerate(category_sales['Purchase Amount (USD)']):
    plt.text(i, v + 100, f"${v:,.0f}", ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

#**Which products are purchased the most, and which ones generate the highest total sales?**

```python
top_product = df['Item Purchased'].value_counts()
top_product
```

```python
top_product2 = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum()
top_product2 = top_product2.sort_values(ascending=False)
top_product2
```

```python
plt.figure(figsize=(20,11))

# The first chart (Number of purchases)
plt.subplot(1, 2, 1)
sns.barplot(
    x=top_product.index,
    y=top_product.values,
    hue=top_product.index,
    palette='viridis'
)
plt.title('Most Purchased Products (Count)', fontsize=14, fontweight='bold')
plt.xlabel('Product',fontsize=15,fontweight='bold')
plt.ylabel('Number of Purchases',fontsize=15,fontweight='bold')
plt.xticks(rotation=45,fontweight='bold')

# The second chart (Total sales of each item)
plt.subplot(1, 2, 2)
sns.barplot(
    x=top_product2.index,
    y=top_product2.values,
    hue=top_product2.index,
    palette='plasma'
)
plt.title('Products by Total Purchase Amount', fontsize=14, fontweight='bold')
plt.xlabel('Product',fontsize=15,fontweight='bold')
plt.ylabel('Total Purchase Amount (USD)',fontsize=15,fontweight='bold')
plt.xticks(rotation=45,fontweight='bold')

plt.tight_layout()
plt.show()
```

#**Which locations have the highest total purchase amounts?**

```python
top_10_locate = df.groupby('Location')['Purchase Amount (USD)'].sum()
top_10_locate = top_10_locate.nlargest(10)
top_10_locate
```

```python
plt.figure(figsize=(15,8))

sns.barplot(
    data=top_10_locate.reset_index(),
    x='Purchase Amount (USD)',
    y='Location',
    hue='Location',
    palette='coolwarm'
)

plt.title('Top 10 Locations by Total Purchase Amount', fontsize=18, fontweight='bold')
plt.xlabel('Total Purchase Amount (USD)', fontsize=14, fontweight='bold')
plt.ylabel('Location', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

# Add values above the columns
for index, value in enumerate(top_10_locate.values):
    plt.text(value + 50, index, f'{value:,.0f}', va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

#**How do total sales vary across different seasons?**

```python
season_sales = df.groupby('Season')['Purchase Amount (USD)'].sum().reset_index()
season_sales
```

```python
plt.figure(figsize=(12,7))

# We arrange them according to the logical sequence of chapters
order = ['Winter', 'Spring', 'Summer', 'Fall']
season_sales['Season'] = pd.Categorical(season_sales['Season'], categories=order, ordered=True)
season_sales = season_sales.sort_values('Season')

sns.lineplot(
    data=season_sales,
    x='Season',
    y='Purchase Amount (USD)',
    marker='o',
    linewidth=3,
    color='#0077b6'
)

plt.title('Total Sales by Season', fontsize=18, fontweight='bold', pad=15)
plt.xlabel('Season', fontsize=14, fontweight='bold')
plt.ylabel('Total Purchase Amount (USD)', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

# Add values above the columns
for i in range(len(season_sales)):
    plt.text(i, season_sales['Purchase Amount (USD)'].iloc[i] + 200,
             f"${season_sales['Purchase Amount (USD)'].iloc[i]:,.0f}",
             ha='center', fontsize=10, fontweight='bold')

sns.despine()
plt.tight_layout()
plt.show()
```

#**Which locations have the highest average product review ratings from customers?**

```python
# We calculate the average rating for each area
avg_rating_per_location = (
    df.groupby('Location')['Review Rating']
      .mean()
      .sort_values(ascending=False)
)

# We display only the first 30 areas if the number is large.
top_avg_rating = avg_rating_per_location.head(30)

plt.figure(figsize=(20,11))
sns.barplot(
    x=top_avg_rating.values,
    y=top_avg_rating.index,
    hue=top_avg_rating.index,
    palette='coolwarm'
)

plt.title('Average Review Rating by Location', fontsize=18, fontweight='bold', pad=15)
plt.xlabel('Average Rating', fontsize=14, fontweight='bold')
plt.ylabel('Location', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

# Add values above the columns
for i, v in enumerate(top_avg_rating.values):
    plt.text(v + 0.02, i, f"{v:.2f}", fontsize=12, fontweight='bold', color='black')

sns.despine()
plt.tight_layout()
plt.show()
```
